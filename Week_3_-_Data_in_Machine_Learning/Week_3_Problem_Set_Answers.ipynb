__1. In a markdown block, describe the role of training testing data in detecting if a machine learning model is overfitting or underfitting.__


In machine learning, training and testing data are crucial for evaluating whether a model is overfitting or underfitting. The training data is used to train the machine learning model. It helps the model learn patterns, relationships, and structures within the data. The testing data is used to evaluate the model's performance on unseen data. It serves as a proxy for how the model will perform in the real world. If the model performs exceptionally well on the training data but poorly on unseen data such as testing data, this is a sign of overfitting. Overfitting occurs when the model memorises the training data instead of generalising it to new data. If the model has low accuracy or high error rates even on the training data, it likely indicates underfitting. Underfitting occurs when the model is too simple to capture the underlying patterns in the data.

__2. In a markdown box, explain 'underfitting' using the below graph as a reference.__

```python
fig, ax = plt.subplots()
x = np.linspace(10, -10, 100)
y = -x**2 + 4*x + 2  
x = x + np.random.uniform(-3, 3, size=x.shape)
y = y + np.random.uniform(-5, 5, size=y.shape)
ax.scatter(x, y, c='b', label='Training & Testing data')
x_lin = np.array([-10,10])
tmp_f_mb = compute_model_output(x_lin, 0, 0,)
ax.plot(x_lin, tmp_f_mb, c='r', label='ML Model')
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_yticks([])
ax.set_yticklabels([])
plt.legend()
plt.show()
```

Underfitting in machine learning occurs when a model is too simplistic to capture the underlying patterns or structure of the data. This typically happens when the model is not complex enough to generalise well with the whole data set. The above training data is non-linear and appears to have a quadratic relationship. The ML Model is a simple linear model that only generalises well with the top of the parabola and generalises very poorly with the lowest points of the parabola.

__3. In a markdown box, explain bias and variance refering to the below model.__ 

```python
fig, ax = plt.subplots()

x = x + np.random.uniform(-3, 3, size=x.shape)
y = y + np.random.uniform(-5, 5, size=y.shape)
ax.scatter(x, y, c='b', label='Training & Testing data')
x = np.linspace(13, -13, 100)
y = -x**2 + 4*x + 2  
ax.plot(x, y, c='r', label='ML Model')
ax.set_xticks([])
ax.set_xticklabels([])
ax.set_yticks([])
ax.set_yticklabels([])
plt.legend()
plt.show()
```

Bias refers to the error introduced by a model when it makes simplifying assumptions about the data. A low-bias model generalises well to the data, meaning it captures the complexities and patterns present in the dataset. The model shows low bias as it follows the quadratic pattern of the data without overfitting or underfitting the data, providing accurate predictions. 

Variance reflects how much a model's predictions change when trained on different subsets of the data. A high-variance model is overly complex, capturing noise as if it were part of the true pattern in the data, which leads to overfitting. This model has generalised to patterns and performed equally well on training data and unseen test data.

The model demonstrates an optimal trade-off to strike a balance between bias and variance.

The model will learn patterns from the training data, which is more distributed with a higher mode and mean and has a more linear relationship and will make predictions from patterns in this data. The test data has less of a linear relationship and a different mode and mean. After the model is trained, if the testing data is used to evaluate it, the model will appear to poorly generalise to the unseeen testing, and The model will struggle to make accurate predictions or perform in real-world applications.
